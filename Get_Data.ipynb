### Những quốc gia không lấy được sẽ bị loại

```python
countries=[]
invalid_countries=['Antarctica','Anguilla','Aruba','Bermuda','Congo','England','Gibraltar','Greenland','Guadeloupe','Guam',\
                  'Guernsey','Jersey','Kiribati','Macao','Martinique','Mayotte','Montserrat','Myanmar','Nauru','Niue',\
                  'Palau','Palestine','Pitcairn','Reunion','Scotland','SriLanka','Swaziland','Saint Pierre and Miquelon',\
                  'South Georgia and the South Sandwich Islands','Tokelau','Tonga','Turkmenistan','Tuvalu',\
                  'United States Minor Outlying Islands','Virgin Islands, U.S.','Wales','Western Sahara',\
                  'Cayman Islands','Cocos (Keeling) Islands','Cook Islands','Ivory Coast','East Timor','Faroe Islands',\
                  'French Polynesia','Holy See (Vatican City State)','New Caledonia','North Korea',\
                  'Northern Mariana Islands','American Samoa','Bouvet Island','British Indian Ocean Territory','Cape Verde',\
                   'Christmas Island','The Democratic Republic of Congo','Czech Republic','Falkland Islands','Fiji Islands',\
                  'French Guiana','French Southern territories','Heard Island and McDonald Islands','Hong Kong','Isle of Man',\
                  'Libyan Arab Jamahiriya','Micronesia, Federated States of','Netherlands Antilles','Norfolk Island',\
                  'Northern Ireland','Puerto Rico','Russian Federation','Saint Helena','South Korea','Svalbard and Jan Mayen',\
                  'Turks and Caicos Islands','United States','Virgin Islands, British','Wallis and Futuna']

for i in range(0,len(a)):
    temp=a[i].get('country')
    if(temp not in invalid_countries):
        countries.append(temp)
```

### Dùng vòng lặp để lấy dữ liệu từng nước 


Quá trình này mất khoảng 2 phút

```python
dataset=[]
for i in range(0,len(countries)):
    url=f'https://covid-api.mmediagroup.fr/v1/cases?country={countries[i]}'
    r=requests.get(url)
    covid_data = json.loads(r.text)
    dataset.append(covid_data['All'])
len(dataset)
```

### Ghi dữ liệu vào File Json

```python
f = open('data\data.json','w')
json.dump(dataset, f)
f.close()
```

### Chuyển file json thành file csv

```python
df=pd.read_json('data\data.json')
df.pop('country')
df.insert(0,'Country',countries)
df.to_csv
df
```
